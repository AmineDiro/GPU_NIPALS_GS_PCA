```python
import numpy as np
import matplotlib.pyplot as plt
from nipals.NIPALS_CPU import *
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

%load_ext autoreload
%autoreload 2
%reload_ext autoreload
```

```python
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal');
```

```python
pipeline = Pipeline([('scaling', StandardScaler()), ('pca', PCA(n_components=1))])

pipeline.fit(X)
X_pca = pipeline.transform(X)

print("original shape:   ", X.shape)
print("transformed shape:", X_pca.shape)
```

```python
std = StandardScaler()
X= std.fit_transform(X)
```

```python
# PCA WITH SVD
u, s, vh = np.linalg.svd(X.T,full_matrices=False)

np.allclose(X.T,  u @ np.diag(s)@vh)

print('Loadings u : ',u.shape)
print('vh shape',vh.shape)

# Compression k=1
Z = u[:,:1].T @ X.T 
Z.shape
```

```python
u @ np.diag(s)
```

```python
nips = Nipals()
nips.fit(X)

print('Loadings P shape : ',nips.loadings.shape)
print('Score T= U @ S shape',nips.scores.shape)
```

```python
Z=nips.transform()
```

```python
T = nips.scores
P = nips.loadings
eig = nips.eig
```

```python
pipeline['pca'].components_
```

```python
U = T/ eig.values
Z = -U.T@ X.T

Z.T.shape
```

```python
error = Z - X_pca 
sum(error > 0.001)
```

## Test Iris dataset 

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target

std =StandardScaler()
X= std.fit_transform(X)
```

```python
n_components = 2
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X)
```

```python
nips = Nipals()
nips.fit(X)
X_nips=nips.transform()

print('Loadings P shape : ',nips.loadings.shape)
print('Score T= U @ S shape',nips.scores.shape)
```

```python
nips = Nipals(ncomp=2)
X_nips = nips.fit_transform(X)
```

```python
for X_transformed, title in [(X_nips, "NIPALS PCA"), (X_pca, "PCA")]:
    plt.figure(figsize=(8, 8))
    for color, i, target_name in zip(colors, [0, 1, 2], iris.target_names):
        plt.scatter(X_transformed[y == i, 0], X_transformed[y == i, 1],
                    color=color, lw=2, label=target_name)

    if "Incremental" in title:
        err = np.abs(np.abs(X_pca) - np.abs(X_ipca)).mean()
        plt.title(title + " of iris dataset\nMean absolute unsigned error "
                  "%.6f" % err)
    else:
        plt.title(title + " of iris dataset")
    plt.legend(loc="best", shadow=False, scatterpoints=1)
    plt.axis([-4, 4, -1.5, 1.5])

plt.show()
```
